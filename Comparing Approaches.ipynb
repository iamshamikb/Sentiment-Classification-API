```python
from matplotlib import pyplot as plt
import seaborn as sns
from IPython.display import clear_output



xgb_tf = [0.87, 0.76, 0.57]
xgb_bow = [0.89, 0.84, 0.58]

svc_tf = [0.91,0.83,0.71]
svc_bow = [0.9,0.78,0.74]

naive_tf = [0.89,0.76,0.71]
naive_bow = [0.89,0.7,0.81]

spacy = [0.89,0.75,0.62]
bert = [0.89,0.9,0.57]



def my_plot(att_name):
    names = ['xgb_tf', 'xgb_bow', 'svc_tf', 'svc_bow', 'naive_tf', 'naive_bow', 'spacy', 'bert']
    att = {'Accuracy':0, 'Precision':1, 'Recall':2}
    
    place = att[att_name]
    values = [xgb_tf[place], xgb_bow[place], svc_tf[place], svc_bow[place], naive_tf[place],\
                        naive_bow[place], spacy[place], bert[place]]

    sns.barplot(names, values)
    plt.title(att_name)

plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
my_plot('Accuracy')
plt.subplot(2,2,2)
my_plot('Precision')
plt.subplot(2,2,3)
my_plot('Recall')
clear_output()
plt.show()
```
